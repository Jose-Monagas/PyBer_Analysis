# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
total_rides_for_cititype = pyber_data_df.groupby(["type"]).count()["ride_id"]
total_rides_for_cititype
```

```python
# 2. Get the total drivers for each city type
total_drivers_for_citytype = city_data_df.groupby(["type"]).sum()['driver_count']
total_drivers_for_citytype
```

```python
#  3. Get the total amount of fares for each city type
total_fares_for_citytype = pyber_data_df.groupby(["type"]).sum()["fare"]
total_fares_for_citytype
```

```python
#  4. Get the average fare per ride for each city type. 
average_fare_per_ride = (total_fares_for_citytype / total_rides_for_cititype)
average_fare_per_ride
```

```python
# 5. Get the average fare per driver for each city type. 
average_fare_per_driver = total_fares_for_citytype / total_drivers_for_citytype
average_fare_per_driver
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.DataFrame({
    "Total Rides": total_rides_for_cititype,
    "Total Drivers": total_drivers_for_citytype,
    "Total Fares": total_fares_for_citytype,
    "Average Fare per Ride": average_fare_per_ride,
    "Average Fare per Driver": average_fare_per_driver
})

pyber_summary_df

```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
```

```python
#  8. Format the columns.
pyber_summary_df["Total Rides"] = pyber_summary_df["Total Rides"].map(float).map("{:,.0f}".format)
pyber_summary_df["Total Drivers"] = pyber_summary_df["Total Drivers"].map(float).map("{:,.0f}".format)
pyber_summary_df["Total Fares"] = pyber_summary_df["Total Fares"].map(float).map("${:,.2f}".format)
pyber_summary_df["Average Fare per Ride"] = pyber_summary_df["Average Fare per Ride"].map(float).map("${:,.2f}".format)
pyber_summary_df["Average Fare per Driver"] = pyber_summary_df["Average Fare per Driver"].map(float).map("${:,.2f}".format)
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
total_fare_date_df = pyber_data_df.groupby(["type","date"]).sum()[["fare"]]
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
total_fare_date_indexed = total_fare_date_df.reset_index()
```

```python
total_fare_date_indexed
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
total_fare_date_pivot = total_fare_date_indexed.pivot(index='date', columns='type', values='fare')
total_fare_date_pivot.head()
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
total_fare_date_pivot = total_fare_date_pivot.reset_index()
is_2019_01_01 = total_fare_date_pivot["date"] >= "2019-01-01"
is_2019_04_29 = total_fare_date_pivot["date"] <= "2019-04-29"
dates_df = total_fare_date_pivot.loc[is_2019_01_01 & is_2019_04_29]
dates_df
```

```python
dates_df.info()
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
dates_df = dates_df.set_index("date")
dates_df.index = pd.to_datetime(dates_df.index)
dates_df
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
dates_df.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
fare_sum_by_week = dates_df.resample("W").sum()
fare_sum_by_week[["Rural","Suburban","Urban"]].head(10)
```

```python
rural = fare_sum_by_week["Rural"]
fare_sum_by_week = fare_sum_by_week.reset_index()
fare_sum_by_week.head()
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 
%matplotlib inline
fig, ax = plt.subplots(figsize=(12, 4))
ax = fare_sum_by_week.plot(x="date", y="Rural", ax=ax)
ax = fare_sum_by_week.plot(x="date", y="Suburban", ax=ax)
ax = fare_sum_by_week.plot(x="date", y="Urban", ax=ax)
ax.set_ylabel("Fare($USD)")
ax.set_title("Total Fare by City Type")
# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')
plt.savefig("analysis/PyBer_fare_summary.png")
plt.show()
```
